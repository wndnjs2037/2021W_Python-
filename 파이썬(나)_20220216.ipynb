```python
# 디테일한 수정 1 - 정확히 한 곳의 값을 변경해주기
df.loc['2019-03-01', 'A'] = 0  #위치를 지정하여 값 바꿔주기 가능
df
```

```python
# 디테일한 수정 2 - 여러 칸을 지정해서 여러 개의 값으로 변경하기 가능
df.loc['2019-03-02', ['B','C']] = 1
df.loc['2019-03-02', ['B','C']] = [1,2] # 1, 2로 입력해도 됨. 값을 각각 변경하려면 리스트의 형태로 값을 대입해준다. 리스트로 덩어리를 넣어줌
df
```

```python
# 디테일한 수정 3 - 2개 이상의 행, 2개 이상의 열의 값 변경하기
df.loc[['2019-03-03','2019-03-04'], ['B','C']] = 0 #값이 여러개가 되는 경우 대괄호로 묶어서 리스트 형태로 전달한다
df
```

```python
# 2개의 행과 2개의 열이므로, 이중 리스트로 만들어서 변경할 값을 전달해주어야 한다.
df.loc[['2019-03-03','2019-03-04'], ['B','C']] = [[1,1],[2,2]] #이중 리스트로 값 전달
df.iloc[[2,3],[1,2]] = [[1,1],[2,2]] #iloc으로도 가능
df
```

```python
# 열 방향 연산
# A에 1을 더한 컬럼 추가하기
df['A+1'] = df['A'] + 1 #새로운 컬럼명을 만들어서 +1을 입력해서 값을 대입해준다.
df
```

```python
# A 열과 B 열을 더하고 싶은 경우
df['A+B'] = df['A'] + df['B']
df
```
